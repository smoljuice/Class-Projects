```python
trips = pd.read_csv('ExData_trips.csv')
trips
```

```python
# How many null values do we have in each column?
trips.isna().sum()
```

```python
# Find rows with null values in start/end station
trips.loc[trips['start_station'].isnull()]
```

```python
# Drop the rows with null values in start/end station
trips = trips.drop(trips.index[[17186, 17323]])
```

```python
trips
```

```python
# Group values by start station
trips.groupby('start_station').count().sort_values(by = 'id')
```

```python
# Create a subset of trips where start and end station were the same
home_trips = trips[trips['start_station'] == trips['end_station']]
home_trips
```

```python
home_trips['start_station'].value_counts()
```

```python
# Create a subset of trips where start station and end station were not the same
com_trips = trips[trips['start_station'] != trips['end_station']]
com_trips
```

```python
com_trips['start_station'].value_counts()
```

```python
# Create an hour column to determine time of day
com_trips['hour'] = pd.to_datetime(com_trips['start_date'], utc = True).dt.hour
```

```python
# create a column indicating day/night
com_trips['is_day'] = [True if 4 <= a <= 15 else False for a in com_trips['hour']]
```

```python
com_trips['is_day'].value_counts()
```
